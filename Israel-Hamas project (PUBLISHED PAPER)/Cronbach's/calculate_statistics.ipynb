# Calculating Cronbach's Alpha



```python
#verify that I did this correctly
def cronbachs(df,ddof=1):
    num_components = len(list(df))
    #print(num_components)
    variances = df.var(ddof=ddof) #variance of each question
    #print(variances)
    var_sum = sum(variances) #sum of all variances
    responder_totals = df.sum(axis=1) #total of each responder
    #print(responder_totals)
    totals_var = responder_totals.var(ddof=ddof) #variance of totals
    alpha = (num_components/(num_components-1))*(1-(var_sum/totals_var))
    return alpha
```
