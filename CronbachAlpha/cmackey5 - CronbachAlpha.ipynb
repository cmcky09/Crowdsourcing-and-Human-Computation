```python
import pingouin as pg
```

```python
genetic.drop([0,1],inplace=True)
neuro.drop([0,1,31],inplace=True)
genetic.reset_index(inplace=True)
neuro.reset_index(inplace=True)
```

```python
danger_genetic_q = genetic[[f'Q{i}' for i in range(2,9)]]
danger_genetic_q = danger_genetic_q.astype('int32')
```

```python
danger_neuro_q = neuro[[f'Q{i}' for i in range(2,9)]]
danger_neuro_q = danger_neuro_q.astype('int32')
```

```python
pg.cronbach_alpha(data=danger_genetic_q)
```

```python
pg.cronbach_alpha(data=danger_neuro_q)
```
